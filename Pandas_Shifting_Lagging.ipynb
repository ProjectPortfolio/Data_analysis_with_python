```python
import pandas as pd
df = pd.read_csv('stock.csv', parse_dates=['Date'], index_col='Date')
df
```

```python
df.shift(1)
#df.shift(-1)
```

```python
df['Prev Day Price'] = df['Price'].shift(1)
df
```

```python
df['Price change'] = df['Price'] - df['Prev Day Price']
df
```

```python
df['4 day change'] = (df['Price'] - df['Price'].shift(4))*100/df['Price'].shift(4)
df
```
