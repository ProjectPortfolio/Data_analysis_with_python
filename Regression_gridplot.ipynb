```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
sns.lmplot(y='tip',x='total_bill',data=df)
```

```python
sns.lmplot(y='tip',x='total_bill',data=df,hue='smoker',markers='+')
```

```python
sns.lmplot(y='tip',x='total_bill',data=df,hue='smoker',markers='D')
```

```python
sns.lmplot(y='tip',x='total_bill',data=df,hue='smoker',markers=['+','o'])
```

```python
sns.lmplot(y='tip',x='total_bill',data=df,hue='smoker',markers=['+','o'],scatter_kws={'s':100})
```

```python
#Regressionplot using GRID
sns.lmplot(x='total_bill',y='tip',data=df,col='smoker')
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,col='smoker',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,col='smoker',row='time',hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,col='day',row='time',hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,col='smoker',hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,col='day',hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,col='smoker',row='ti,hue='sex',aspect=0.5)
```
