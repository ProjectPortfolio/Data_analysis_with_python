```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('weather_data.csv',parse_dates=['day'])
df
```

```python
newdf = df.set_index('day') #set_index() is a method to set a List, Series or Data frame as index of a Data Frame.
newdf
```

```python
newdf.interpolate(method="time") #dataframe.interpolate() function is basically used to fill NA values in the 
#dataframe or series. But, this is a very powerful function to fill the missing values.
#Syntax: DataFrame.interpolate(method=’linear’, axis=0, limit=None, 
#inplace=False, limit_direction=’forward’, limit_area=None, downcast=None, **kwargs)
```

```python
df.dropna() #Pandas dropna() method allows the user to analyze and drop Rows/Columns with Null 
#values in different ways.
```

```python
df.dropna(how ='all')
newdf.dropna(how='all')
#how: how takes string value of two kinds only (‘any’ or ‘all’).
#‘any’ drops the row/column if ANY value is Null and ‘all’ drops only if ALL values are null
```

```python
#Creating Pandas DataFrame from lists of lists.
weather_data = [["1/1/2017",32,6,'Rain'],
["1/2/2017",35,7,'Sunny'],
["1/3/2017",28,7,'Snow']]
df = pd.DataFrame(weather_data,
                 columns = ["day","temperature","windspeed","event"])
df
```

```python
# Creating Dataframe from list of dicts

weather_data=[{'day':'1/01/2017','temperature':32,'windspeed':6,'event':'Rain'},
             {'day':'1/02/2017','temperature':35,'windspeed':7,'event':'Sunny'},
             {'day':'1/02/2017','temperature':28,'windspeed':8,'event':'snow'}]
df=pd.DataFrame(weather_data)
df
```
