```python
import seaborn as sns
df = sns.load_dataset('flights')
df.head()
```

```python
sns.relplot(x='month', y='passengers', data=df, hue='year', kind='line')
```

```python
sns.catplot(x='month', y='passengers', data=df, kind='box')
```

```python
df2 = sns.load_dataset('tips')
sns.distplot(df2['total_bill'])
```

```python
sns.jointplot(x='total_bill', y='tip', data=df2, kind='reg')
```

```python
sns.pairplot(df2)
```
