```python
import matplotlib.pyplot as plt
import numpy
```

```python
#Line plotting
x = [1, 2, 3,4 ,5, 6, 7]
y = [30, 34, 25, 45, 32, 36, 29]

plt.plot(x, y, color='b', linewidth=4, marker="+")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Daywise Temp")
```

```python
plt.plot(x, y, 'r+-.', alpha=0.2)
plt.plot([1, 2, 3, 4], [10, 20, 30, 40])
```

```python
days = [1, 2, 3, 4, 5, 6, 7]
max_t = [50, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t = [45, 48, 47, 46, 40, 42, 41]

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")
plt.grid()
plt.legend(loc="best", shadow=True)
```

```python
#Bar Charts
company = ['FB', 'MSFT', 'AMZN', 'APPLE']
revenue = [80, 100, 79, 17]
profit = [40, 2, 34, 12]

plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Stock Analysis")
plt.bar(company, revenue, width=0.4, label="Revenue")
plt.bar(company, profit, width=0.2, label="Profit")

#plt.barh(company, revenue, label="Revenue")
#plt.barh(company, profit, label="Profit")

plt.legend()
```
