```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://www.monster.com/jobs/search/?q=Software-Developer&where=Indiana"
```

```python
page = requests.get(URL)
#page.content
```

```python
soup = BeautifulSoup(page.content, 'html.parser')

results = soup.find(id='ResultsContainer')
```

```python
#results.prettify()
jobs = results.find_all('section', class_='card-content')
```

```python
title = []
company = []
loc = []
for job in jobs:
    job_title = job.find('h2', class_='title')
    company_name = job.find('div', class_='company')
    location = job.find('div', class_='location')
    
    if None in (job_title, company_name, location):
        continue
    
    #print(job_title.text)
    #print(company_name.text)
    #print(location.text)
    title.append(job_title.text.strip())
    company.append(company_name.text.strip())
    loc.append(location.text.strip())
```

```python
import pandas as pd

data = {
    'JobTitle' : title,
    'CompanyName' : company,
    'Location' : loc
}

df = pd.DataFrame(data)
df
```
